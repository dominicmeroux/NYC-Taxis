# GraphFrames Analysis

Unless otherwise noted, code in many cases has been adapted from https://docs.databricks.com/_static/notebooks/graph-analysis-graphframes.html to fit this application

Ran this section individually for Chicago, then Yellow Cab

```
# Obtain distinct rounded trip start and end locations
edges = (DF.select(DF['src'].alias('id')).union(DF.select(DF['dst'].alias('id')))).distinct()

display(edges)
```

```
# Create Graph! 
# Useful resources:
# Databricks Example Notebook Using Bikeshare: https://docs.databricks.com/_static/notebooks/graph-analysis-graphframes.html
# GraphFrames User Guide http://graphframes.github.io/user-guide.html#pagerank
# IBM PageRank https://developer.ibm.com/clouddataservices/2016/07/15/intro-to-apache-spark-graphframes/
# Conceptual PowerPoint destination http://www.cs.columbia.edu/~lierranli/publications/GraphFrames-SparkSummit2016.pdf
g = GraphFrame(edges, DF)
```

#### PageRank Algorithm

```
# PageRank Algorithm
ranks = g.pageRank(resetProbability=0.15, maxIter=10)

display(ranks.vertices.select("id","pagerank").orderBy(desc("pagerank")))
```

#### Most Common Trips

```
# Most Common Trips
topTrips = g.edges.groupBy("src", "dst").count().orderBy(desc("count"))#.limit(10)

display(topTrips)
```

#### In-Degree

```
inDeg = g.inDegrees
display(inDeg.orderBy(desc("inDegree")))#.limit(10))
```

#### Out-Degree

```
outDeg = g.outDegrees
display(outDeg.orderBy(desc("outDegree")))#.limit(10))
```

#### Locations with highest traffic as a destination, lowest traffic out

```
degreeRatio = inDeg.join(outDeg, inDeg["id"] == outDeg["id"]).drop(outDeg["id"]).selectExpr("id", "double(inDegree)/double(outDegree) as degreeRatio")

degreeRatio.cache()
  
display(degreeRatio.orderBy(desc("degreeRatio")))#.limit(10))
```

#### Locations with highest traffic as an origin, lowest traffic in

```
display(degreeRatio.orderBy(asc("degreeRatio")))#.limit(10))
```

#### Label Propagation Algorithm to Detect Communities

Code modified from IBM approach https://developer.ibm.com/clouddataservices/2016/07/15/intro-to-apache-spark-graphframes/

```
communities = g.labelPropagation(maxIter=5).select("id", "label").groupBy("label").agg(count("id").alias("count")).orderBy(desc("count"))

display(communities)
```
